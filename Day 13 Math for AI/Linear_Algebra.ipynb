## Coordinate Geometry & Straight Lines

### Trigonometry Basics
* In a right-angled triangle with Hypotenuse ($H$), Perpendicular ($P$), and Base ($B$):
    * $\sin\theta = \frac{P}{H}$
    * $\cos\theta = \frac{B}{H}$
    * $\tan\theta = \frac{P}{B}$

### Equation for a Straight Line
$$y = mx + c$$
* **$m$** = slope = $\tan\theta$
* **$c$** = intercept (where the line crosses the y-axis, point $(0, c)$)

**Finding Slope ($m$) from Two Points:**
Given $P_1(x_1, y_1)$ and $P_2(x_2, y_2)$:
$$m = \tan\theta = \frac{P}{B} = \frac{y_2 - y_1}{x_2 - x_1}$$

**Using a point and the intercept:**
Given a point $(x_1, y_1)$ and intercept $(0, c)$:
$$m = \frac{y_1 - c}{x_1 - 0} \implies m = \frac{y_1 - c}{x_1}$$

## Distance Formulas & Line Properties

### Distance Between 2 Points
* **2D Space:** (derived from Pythagorean theorem $H^2 = B^2 + P^2$)
    $$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$
* **3D Space:**
    $$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$$

### Parallel and Perpendicular Lines
* **Parallel Lines:**
    * $L_1: y = mx + c_1$
    * $L_2: y = mx + c_2$
    * **Rule:** $m_1 = m_2 = m$
* **Perpendicular Lines:**
    * $L_1: y = m_1x + c_1$
    * $L_2 \text{ slope } (m_2) = \tan(90^\circ + \theta) = \frac{\sin(90^\circ + \theta)}{\cos(90^\circ + \theta)} = \frac{\cos\theta}{-\sin\theta} = -\frac{1}{\tan\theta}$
    * **Rule:** $m_2 = -\frac{1}{m_1} \implies m_1 \cdot m_2 = -1$
    * $L_2: y = -\frac{1}{m_1}x + c_2$

## Distance Between Lines

### Distance Between 2 Parallel Lines
Consider parallel lines $L_1: y = mx + c_1$ and $L_2: y = mx + c_2$.
Looking at the vertical distance along the y-axis, we use trigonometric relationships where the angle between the vertical and the perpendicular distance $d$ is $\theta$:
$$\cos\theta = \frac{d}{|c_2 - c_1|} \implies d = |c_2 - c_1|\cos\theta$$

To find $\cos\theta$ in terms of slope $m$:
$$m = \tan\theta = \frac{\sin\theta}{\cos\theta}$$
$$m^2 = \frac{\sin^2\theta}{\cos^2\theta} = \frac{1 - \cos^2\theta}{\cos^2\theta}$$
$$m^2 \cos^2\theta = 1 - \cos^2\theta$$
$$\cos^2\theta(m^2 + 1) = 1 \implies \cos^2\theta = \frac{1}{1 + m^2} \implies \cos\theta = \frac{1}{\sqrt{1 + m^2}}$$

Substituting back:
$$d = \frac{|c_2 - c_1|}{\sqrt{1 + m^2}} \quad \text{(Important Formula)}$$

### Practice: Distance of a Point to a Line
**What is the distance of a point $P(a,b)$ placed perpendicular to a Line ($y=mx+c$)?**
* Original Line $L_1: y = mx + c$
* Construct a parallel line $L_2$ passing through $P(a,b)$: $y = mx + c'$
    $$b = ma + c' \implies c' = b - am$$
* Using the distance formula between parallel lines:
    $$d = \frac{|c' - c|}{\sqrt{1 + m^2}} \implies d = \frac{|(b - am) - c|}{\sqrt{1 + m^2}}$$
    $$d = \frac{|b - am - c|}{\sqrt{1 + m^2}}$$

## Vectors

A vector is a quantity that has both magnitude (size/length) and direction. Often represented as an arrow ($\vec{v}$).

* **Direction:** represented by unit components on axes $x, y, z$.
    $$\vec{V} = V_1\hat{i} + V_2\hat{j} + V_3\hat{k}$$
* **Magnitude:** Distance from origin $(0,0,0)$.
    $$|\vec{V}| = \sqrt{V_1^2 + V_2^2 + V_3^2}$$

### Vector Operations
**1) Vector Addition:**
Given $A = (2, 1, 3)$ and $B = (4, 5, 2)$:
$$\vec{A} = 2\hat{i} + 1\hat{j} + 3\hat{k}$$
$$\vec{B} = 4\hat{i} + 5\hat{j} + 2\hat{k}$$
$$\vec{A} + \vec{B} = 6\hat{i} + 6\hat{j} + 5\hat{k}$$

**2) Scalar Multiplication:**
Given $A = (2, 1, 3) \times 3$:
$$\vec{A} = (2\hat{i} + 1\hat{j} + 3\hat{k}) \times 3$$
$$\vec{A} = 6\hat{i} + 3\hat{j} + 9\hat{k}$$

## Vector Multiplication (Dot & Cross Product)

### 3) Dot Product (Scalar Product)
$$\vec{A} = a_1\hat{i} + b_1\hat{j} + c_1\hat{k}$$
$$\vec{B} = a_2\hat{i} + b_2\hat{j} + c_2\hat{k}$$
$$\vec{A} \cdot \vec{B} = a_1a_2 + b_1b_2 + c_1c_2 = |\vec{A}||\vec{B}|\cos\theta$$
*Example:* For $A=(2,1,3)$ and $B=(4,5,2)$
$$\vec{A} \cdot \vec{B} = (2\times4) + (1\times5) + (3\times2) = 8 + 5 + 6 = 19$$

**Projection of one vector onto another:**
* $\cos\theta = \frac{B}{H} = \frac{\text{Proj of A}}{|\vec{A}|}$
* $\text{Proj of A on B} = |\vec{A}|\cos\theta$
* Multiply by $\frac{|\vec{B}|}{|\vec{B}|}$:
    $$\text{Proj of A on B} = \frac{|\vec{A}||\vec{B}|\cos\theta}{|\vec{B}|} = \frac{\vec{A} \cdot \vec{B}}{|\vec{B}|}$$

### 4) Cross Product (Vector Product)
It produces a new vector that is perpendicular to both original vectors.
$$|\vec{A} \times \vec{B}| = |\vec{A}||\vec{B}|\sin\theta$$
*(Uses right-hand rule to determine if the new vector goes out of the plane $\odot$ or into the plane $\otimes$)*

## Matrices

A matrix (plural: matrices) is a rectangular array of numbers, symbols, or expressions arranged in rows and columns.

### Operations:
* **Matrix Addition/Subtraction ($A \pm B$):** Dimensions must match ($m \times n \pm m \times n$).
* **Scalar Multiplication ($n \cdot A$):** $\lambda A$
* **Matrix Multiplication ($A \times B$):** Cols of A must equal Rows of B ($m \times n \text{ and } p \times q \implies n=p$). The resulting matrix size is $m \times q$.
* **Dot Product for Row and Col Matrix**
* **Transpose ($A^T$)**

**Multiplication Example ($A \times B$):**
$$A = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix}, \quad B = \begin{bmatrix} 8 & 3 & 1 \\ 4 & 2 & 3 \end{bmatrix}$$
$$A \times B = \begin{bmatrix} (8+12) & (3+6) & (1+9) \\ (16+16) & (6+8) & (2+12) \end{bmatrix} = \begin{bmatrix} 20 & 9 & 10 \\ 32 & 14 & 14 \end{bmatrix}$$

**Dot Product Example:**
$$R = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}_{1 \times n}, \quad C = \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix}_{n \times 1}$$
$$R \cdot C = [4 + 10 + 18] = 32$$

## Determinants

A determinant ($|A|$) is a single number calculated from a square matrix.

**2x2 Matrix:**
$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \implies |A| = ad - cb$$

**3x3 Matrix:**
$$A = \begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix}$$
$$|A| = a \cdot M_{11} - b \cdot M_{12} + c \cdot M_{13}$$
* Where $M_{11} = (ei - hf)$, $M_{12} = (di - gf)$, $M_{13} = (dh - ge)$
* **Co-factor rule:** $C_{ij} = (-1)^{i+j} M_{ij}$ (Even positions are +, odd are -).

**Geometric Interpretation of Linear Equations:**
* **2D:**
    $L_1: a_1x + b_1y = c_1$
    $L_2: a_2x + b_2y = c_2$
    If $\begin{vmatrix} a_1 & b_1 \\ a_2 & b_2 \end{vmatrix} \neq 0$, there is a **unique solution** (the lines are not parallel and intersect).
* **3D:**
    If $\begin{vmatrix} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{vmatrix} \neq 0$, the planes intersect uniquely (not parallel).

## Eigenvectors & Eigenvalues

An **eigenvector** of a matrix is a non-zero vector that, when the matrix acts on it, does not change direction—it only gets scaled.
$$AV = \lambda V$$
* **$A$** = square matrix
* **$V$** = eigenvector
* **$\lambda$** = eigenvalue (scaling factor)

### Finding Eigenvalues and Eigenvectors
**Example:** Let $A = \begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix}$

**Step 1: Set up the Characteristic Equation**
$$AV = \lambda V$$
$$AV - \lambda V = 0$$
$$(A - \lambda I)V = 0$$
* $V$ is not zero because it has direction.
* $I$ is the identity matrix $\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$.

To find non-trivial solutions, we set the determinant to zero:
$$|A - \lambda I| = 0$$

**Step 2: Solve for $\lambda$ (Eigenvalues)**
$$\begin{vmatrix} 4-\lambda & 1 \\ 2 & 3-\lambda \end{vmatrix} = 0$$
$$(4-\lambda)(3-\lambda) - (2)(1) = 0$$
$$12 - 4\lambda - 3\lambda + \lambda^2 - 2 = 0$$
$$10 - 7\lambda + \lambda^2 = 0$$

Factor the quadratic equation:
$$10 - 5\lambda - 2\lambda + \lambda^2 = 0$$
$$5(2-\lambda) - \lambda(2-\lambda) = 0$$
$$(2-\lambda)(5-\lambda) = 0$$
**Eigenvalues:** $\lambda = 5, \lambda = 2$

**Step 3: Solve for $V$ (Eigenvector) using $\lambda=5$**
Substitute $\lambda=5$ back into $(A - \lambda I)V = 0$:
$$\begin{bmatrix} 4-5 & 1 \\ 2 & 3-5 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$
$$\begin{bmatrix} -1 & 1 \\ 2 & -2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

This gives the system of equations:
$$-x + y = 0 \implies x = y$$
$$2x - 2y = 0 \implies x = y$$

So, for $\lambda = 5$, the eigenvector is:
$$V = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$$
