## SVR - regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
df= datasets.load_diabetes(as_frame=True).frame
df.head()
```

```python
X=df.drop(columns="target",axis=1)
y=df["target" ]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scalar=StandardScaler()
y_train_scaled=scalar.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_scaled=scalar.transform(y_test.values.reshape(-1,1)).ravel()
```

```python
svr=SVR(kernel="rbf")
svr.fit(X_train,y_train_scaled)
```

```python
y_test_pred_scaled=svr.predict(X_test)
y_train_pred_scaled=svr.predict(X_train)
```

```python
print("r2 : ",r2_score(y_train_scaled,y_train_pred_scaled))
print("r2 : ",r2_score(y_test_scaled,y_test_pred_scaled))
```

```python
# Linear Kernel
svr=SVR(kernel="linear")
svr.fit(X_train,y_train_scaled)
y_test_pred_scaled=svr.predict(X_test)
y_train_pred_scaled=svr.predict(X_train)
print("r2 : ",r2_score(y_train_scaled,y_train_pred_scaled))
print("r2 : ",r2_score(y_test_scaled,y_test_pred_scaled))
```

```python
# ploy Kernel
svr=SVR(kernel="poly")
svr.fit(X_train,y_train_scaled)
y_test_pred_scaled=svr.predict(X_test)
y_train_pred_scaled=svr.predict(X_train)
print("r2 : ",r2_score(y_train_scaled,y_train_pred_scaled))
print("r2 : ",r2_score(y_test_scaled,y_test_pred_scaled))
```

```python
# Sigmoid Kernel
svr=SVR(kernel="sigmoid")
svr.fit(X_train,y_train_scaled)
y_test_pred_scaled=svr.predict(X_test)
y_train_pred_scaled=svr.predict(X_train)
print("r2 : ",r2_score(y_train_scaled,y_train_pred_scaled))
print("r2 : ",r2_score(y_test_scaled,y_test_pred_scaled))
```

# Hyperparmeter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "C":[1,2,5,10,50,100],
    "kernel": ["rbf", "linear", "poly", "sigmoid"],
    "epsilon":[0.01,0.1,0.2,0.3,0.5]
}
```

```python
svr=SVR()
grid_search=GridSearchCV(svr,param_grid=param_grid,scoring='r2',cv=5)
grid_search.fit(X_train, y_train_scaled)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```
