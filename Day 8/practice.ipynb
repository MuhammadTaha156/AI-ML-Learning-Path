```python
import json
```

```python
def load_data(data):
    with open(data,"r") as f:
        data=json.load(f)

        return data
```

```python
data=load_data("store_data.json")
data
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    clean_Data=[]
    unique_users=set()

    for user in data:
        # Clean Rating
        if str(user["rating"]).strip().lower() in text_to_num:
            user["rating"]=text_to_num[user["rating"]]
        
        # Handling Missing Values
        if user.get("age")==None:
            user["age"]=None

        # Deduplication
        if user["name"].strip() in unique_users:
            continue
        unique_users.add(user["name"])
        clean_Data.append(user)

    
    return clean_Data

```

```python
cleanData=clean_data(data)
cleanData
```

```python
def get_insight(data):
     total_rating=0
    #  Avg Rating
     for user in data:
          total_rating+=float(user["rating"])
     avg_rating=total_rating/len(data)
     print(f"Average of Rating : {avg_rating}")

     # % of Users with poor Rating
     poor_rating=0
     for user in data:
          if float(user["rating"])<3:
               poor_rating+=1
     print(f"% of Users gave poor Rating : {(poor_rating/len(data))*100}")
          
```

```python
get_insight(cleanData)
```

```python
# Bulid a Recommandation Feature
def recommandations(data):
    recommandation=[]
    for user in data:
        curr_recom={}
        if float(user["rating"])>=4:
            curr_recom={"Name":user["name"],"Brand":"Apple"}
        else:
            curr_recom={"Name":user["name"],"Brand":"Sumsang"}
        recommandation.append(curr_recom)
    return recommandation


```

```python
recommandations(cleanData)
```
