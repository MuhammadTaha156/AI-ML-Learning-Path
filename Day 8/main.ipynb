```python
import json
```

```python
#  Load the Data
def load_Data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    
    return data
```

```python
data=load_Data("store_data.json")
type(data)
data
```

```python
# Cleam and Structure the Data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    clean_data=[]
    unique_users=set()
    for user in data:
        # Clean Ratings
        raw_rating=str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]

        user["rating"]=raw_rating


        # Handle Missing Values
        raw_age=user.get("age")
        if raw_age==None:
            user["age"]=None

        # Deduplication
        if (user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        clean_data.append(user)
    
    return clean_data
```

```python
cleanData=clean_data(data)
```

```python
#  Get Meaningful insights from data
def get_insights(data):
    # avg rating
    total_rating=0
    for user in data:
        total_rating+= float(user["rating"])

    print(f"Avg Rating = {total_rating/len(data)}")

    # % of Users with poor Rating
    poor_rating=0
    for user in data:
        if float(user["rating"])<3:
            poor_rating+=1
     
    print(f"% of user with Poor Rating = {poor_rating/len(data)*100}")
```

```python
get_insights(cleanData)
```

```python
#  Build Recommendation Feature
def get_recommandations(data):
    recommendation=[]
    for user in data:
        cur_recomm={}
        cur_recomm["name"]=user["name"]
        if float(user["rating"])>=4:
            cur_recomm["brand"]="Apple"
        else:
            cur_recomm["brand"]="Samsung"
        recommendation.append(cur_recomm)
    return recommendation
```

```python
get_recommandations(cleanData)
```
